## Normalizing the transforming features with MinMaxScalar() and fit_transform()

```python
address = r'C:\Users\user\Desktop\DataScienceTutorial\DataSets\mtcars.csv'
cars = pd.read_csv(address)
cars.head()
```

```python
mpg = cars.mpg
plt.plot(mpg)
```

```python
cars[['mpg']].describe()
```

```python
mpg_matrix = mpg.values.reshape(-1,1)

scaled = preprocessing.MinMaxScaler()

scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
scaled = preprocessing.MinMaxScaler(feature_range=(0,10))

scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

## Using Scale() to scale your features

```python
standarized_mpg = scale(mpg, axis=0, with_mean=False, with_std=False)
plt.plot(standarized_mpg)
```
